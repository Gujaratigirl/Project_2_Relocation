```python
# Import Dependencies
import pandas as pd
import os
import psycopg2
from sqlalchemy import create_engine
```

```python
file_path = os.path.join("Data", "state_migrations_2019.xlsx")
df = pd.read_excel(file_path, header=0, index_col="Current State", sheet_name="data")
df
```

```python
states = ["California", "Florida", "Indiana", "North Carolina", "Texas"]

CA_in_ser = df.iloc[4, :]
FL_in_ser = df.iloc[9, :]
IN_in_ser = df.iloc[14, :]
NC_in_ser = df.iloc[33, :]
TX_in_ser = df.iloc[43, :]
```

```python
CA_in_ser
```

```python
CA_in_df = CA_in_ser.to_frame(name="IN")
FL_in_df = FL_in_ser.to_frame(name="IN")
IN_in_df = IN_in_ser.to_frame(name="IN")
NC_in_df = NC_in_ser.to_frame(name="IN")
TX_in_df = TX_in_ser.to_frame(name="IN")
```

```python
CA_in_df
```

```python
CA_in_df["State"] = "California"
FL_in_df["State"] = "Florida"
IN_in_df["State"] = "Indiana"
NC_in_df["State"] = "North Carolina"
TX_in_df["State"] = "Texas"
```

```python
CA_in_df
```

```python
CA_out_ser = df.iloc[:, 4]
FL_out_ser = df.iloc[:, 9]
IN_out_ser = df.iloc[:, 14]
NC_out_ser = df.iloc[:, 33]
TX_out_ser = df.iloc[:, 43]
```

```python
CA_out_ser
```

```python
CA_out_df = CA_out_ser.to_frame(name="OUT")
FL_out_df = FL_out_ser.to_frame(name="OUT")
IN_out_df = IN_out_ser.to_frame(name="OUT")
NC_out_df = NC_out_ser.to_frame(name="OUT")
TX_out_df = TX_out_ser.to_frame(name="OUT")
```

```python
CA_out_df
```

```python
CA_out_df["State"] = "California"
FL_out_df["State"] = "Florida"
IN_out_df["State"] = "Indiana"
NC_out_df["State"] = "North Carolina"
TX_out_df["State"] = "Texas"
```

```python
CA_out_df
```

```python
CA_merge_df = CA_in_df.merge(CA_out_df, left_index=True, right_index=True)
FL_merge_df = FL_in_df.merge(FL_out_df, left_index=True, right_index=True)
IN_merge_df = IN_in_df.merge(IN_out_df, left_index=True, right_index=True)
NC_merge_df = NC_in_df.merge(NC_out_df, left_index=True, right_index=True)
TX_merge_df = TX_in_df.merge(TX_out_df, left_index=True, right_index=True)
```

```python
CA_merge_df
```

```python
append_two_df = CA_merge_df.append(FL_merge_df)
append_three_df = append_two_df.append(IN_merge_df)
append_four_df = append_three_df.append(NC_merge_df)
append_five_df = append_four_df.append(TX_merge_df)
```

```python
index_move_df = append_five_df.reset_index()
index_move_df
```

```python
drop_df = index_move_df.drop(columns="State_y")
rename_df = drop_df.rename(columns={"index": "secondary_state", "State_x": "primary_state", "IN": "inflow", "OUT": "outflow"})
```

```python
final_df = rename_df[["primary_state", "secondary_state", "inflow", "outflow"]]
final_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/migration_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
final_df.to_sql(name='relocation', con=engine, if_exists='append', index=False)
```
